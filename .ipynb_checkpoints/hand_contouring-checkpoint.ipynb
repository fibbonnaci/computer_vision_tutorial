### Import resources and display image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
# read in the image
image = cv2.imread('images/raised_hand.png')

# make a copy
image_copy = np.copy(image)

# change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

### Produce a binary image for finding contours

```python
# convert to gray scale
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# create a binary threshold image
retval, binary = cv2.threshold(gray, 250, 250, cv2.THRESH_BINARY_INV)

plt.imshow(binary, cmap='gray')
```

```python
# find contours from thresholded image
retval, contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

print(len(contours))
print(contours[0])
print(contours[1])
```

```python
# draw all contours on a copy of the original image
copy = np.copy(image_copy)
all_contours = cv2.drawContours(copy, contours, -1, (0,255,0), 2) # using -1 means display all contours

plt.imshow(all_contours)
```

### Try another image

```python
# read in the image
up_down = cv2.imread('images/thumbs-up-down.jpg')

# change color to RGB (from BGR)
up_down = cv2.cvtColor(up_down, cv2.COLOR_BGR2RGB)

plt.imshow(up_down)
```

```python
# convert to gray scale
gray = cv2.cvtColor(up_down, cv2.COLOR_RGB2GRAY)

# create a binary threshold image
retval, binary = cv2.threshold(gray, 215, 170, cv2.THRESH_BINARY_INV) # play with the thresh val up and down

# find contours from thresholded image
retval, contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

print(len(contours))
# print(contours[0])
# print(contours[1])
```

```python
# Select the contour at index= 0 from a list
selected_contour = contours[1]

# draw the first contour (index = 0)
contour_image = np.copy(up_down)
contour_image = cv2.drawContours(contour_image, [selected_contour], -1,  (0,255,0), 2)

plt.imshow(contour_image)
```
