### Import resources and display image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
# read in the image
image = cv2.imread('images/sunflower.jpg')

# make a copy
image_copy = np.copy(image)

# change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
# convert to grayscale for filtering
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

### Implement Canny edge detection

```python
# define lower and upper threshold for hysteresis 
# it is recommended to use lower:upper value ratio of 1:2 or 1:30
lower = 120
upper = 240

edges = cv2.Canny(gray, lower, upper)

plt.imshow(edges, cmap='gray')
```

```python
# Now, try Canny again using "wide" and "tight" thresholds

wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 180, 240)

# create side-by-side plots
f, (ax1, ax2) = plt.subplots(1,2, figsize=(40,60))

ax1.imshow(wide, cmap='gray')

plt.imshow(tight, cmap='gray')
```
